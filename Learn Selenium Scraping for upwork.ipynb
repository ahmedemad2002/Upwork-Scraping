```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import matplotlib.pyplot as plt
import os
import logging
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
page_url = "https://v2.hklii.hk/en/legis/ord/"

driver.get(page_url)
time.sleep(3)

cap_no = []
title = []
links = []
next_page_btn = driver.find_element("xpath", "//button[@type='button' and @class='v-btn v-btn--flat v-btn--icon v-btn--round v-btn--text theme--light v-size--default']")

for i in range(8):
    time.sleep(2)
    rows = driver.find_elements(By.CLASS_NAME, "routing")
    r_text = [row.text for row in rows]
    
    cap_no.extend(r_text[i] for i in range(0, len(r_text), 2))
    title.extend(r_text[i] for i in range(1, len(r_text), 2))
    for i in range(0, len(rows), 2):
        links.append(str(rows[i].get_attribute('href')))
    next_page_btn.click()
```

```python
df = pd.DataFrame({'id': cap_no, 'Title': title, 'Link': links})
```
